```python
def activation(out,threshold):
  if out>=threshold:
    return 1
  else:
    return 0
```

```python
import numpy as np
andInput = np.array([
    [1,0],
    [1,1],
    [0,0],
    [0,1]
    ])

# Nesse caso ele está passando por cada casa do "a" e do "b" e ao passar ele vê o quanto
# errou e corrige cada peso
def perceptron(andInput):
  a = [1,1,0,0]
  b = [1,0,1,0]
  x = [1,0,0,0] # Output desejado
  weights = [0.1,0.8] # peso
  bias = 0.2
  threshold = 1
  lr = 0.5
  i = 0
  #colocando as epocas
  epochs = 0
  while epochs <= 10000:
    i = 0

    while i < 4:
        soma = a[i]*weights[0] + b[i]*weights[1] + bias
      #activationFunc
        out = activation(soma, threshold)
      #ajuste se teve um erro
        if (out != x[i]):
            error = x[i] - out
            weights[0] = weights[0] + lr * error * a[i]
            weights[1] = weights[1] + lr * error * b[i]
            bias = bias + lr * error
        i = i + 1
    epochs = epochs + 1
  print('\n')
#mostrando ao usuario o resultado com esse peso obtido
  i = 0
  while i < 4:
    soma = np.dot(andInput[i],weights) + bias
    out = activation(soma, threshold)
    print(str(andInput[i]) + ' = ' + str(out))
    i = i + 1

  print('Pesos obtidos: '+ str(weights))
  print('Bias obtido: '+ str(bias))

perceptron(andInput)
```
